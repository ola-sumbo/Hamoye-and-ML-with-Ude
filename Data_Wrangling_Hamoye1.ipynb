```
import pandas as pd
import numpy as np
cell_data = pd.read_csv('https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/cell_phones.csv')
pop_data = pd.read_csv('https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/population1.csv')
geo_code = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv')
```

**Joining the data**

```
merged = cell_data.merge(pop_data, how='inner', on = ['Country', 'year'])
merged.shape
```

```
merged.head()
```

```
# renaming columns in geodataframe
geo_codes = geo_code.rename(columns={'country':'geo', 'name':'Country'})
geo_code.head()
```
