```
import mpl_toolkits.basemap
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(12, 9))
map = Basemap(projection ='mill', 
              llcrnrlat=-90,
              urcrnrlat=90,
              urcrnrlon=180,
              llcrnrlon=-180, resolution='c') # when resolution = c it is crude, l = low , i = intermediate, h = high, f = full resolution 
map.drawcoastlines()
#map.drawcountries(color='red') #draws country boundaries
#map.drawstates(color='blue')
#map.drawcounties(color='orange')
#map.drawrivers(color='blue')

#map.drawmapboundary(color ='purple', linewidth = 30, fill_color='aqua')
#map.fillcontinents(color='lightgreen', lake_color='aqua')
#map.drawlsmask(land_color='lightgreen', ocean_color='aqua', lakes=True)
#map.etopo() #shows maps topography
#map.bluemarble()
#map.shadedrelief()

#draw grid lines
map.drawparallels(np.arange(-90, 90, 10), labels=[True,False, False, False])
map.drawmeridians(np.arange(-180, 180, 30), labels=[0,0,0,1])
#np.arange(start,stop,step)
#labels=[left,right,top,bottom]

plt.title('Basemap Beginner', fontsize=30)
plt.show()
```

```
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

map = Basemap(projection='ortho', 
              lat_0=0, lon_0=0)

map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral',lake_color='aqua')
map.drawcoastlines()

x, y = map(0, 0)

map.plot(x, y, marker='D',color='m')

plt.show()
```

```
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

map = Basemap(projection='ortho', 
              lat_0=0, lon_0=0)

map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral',lake_color='aqua')
map.drawcoastlines()

lons = [0, 10, -20, -30]
lats = [0, -10, 40, -20]

x, y = map(lons, lats)

map.scatter(x, y, marker='D',color='m')

plt.show()
```
